**1.** Durante a aula, discutimos como construir uma priori GP e o formato da posteriori preditiva para problemas de regressão com verossimilhança Gaussiana (com média definida pelo GP). O código abaixo cria um GP com kernel exponencial quadrático, mostra a priori preditiva e a posteriori preditiva. Experimente com o código e comente a influência de ambos os parâmetros do kernel exponencial quadrático, tanto na priori preditiva quanto na posteriori preditiva. Nos gráficos gerados, os pontos vermelhos são observações, as curvas sólidas azuis são a médias das preditivas e o sombreado denota +- um desvio padrão.

```python
SEED = 42
np.random.seed(SEED)

s2 = 1e-04 # variância observacional

def rbf_kernel(x1, x2, gamma=10.0, c=1.0):
    assert(gamma>0)
    assert(c>0)
    return (-gamma*(torch.cdist(x1, x2)**2)).exp()*c

x = torch.linspace(-1, 1, 100)[:, None]

K = rbf_kernel(x, x) + torch.eye(x.shape[0])*s2
mu = torch.zeros_like(x)

fig, axs = plt.subplots(8,4, figsize=(12, 24))



xtrain = torch.tensor([-0.5, 0.0, 0.75])[:, None]
ytrain = torch.tensor([-1.5, 1.0, 0.5])[:, None]

def posterior_pred(x, xt, yt, gamma=10.0, c=1.0):
    Kxxt = rbf_kernel(x, xt, gamma, c)
    Kxt = rbf_kernel(xt, xt, gamma, c) + torch.eye(xt.shape[0])*s2
    Kinv = torch.linalg.inv(Kxt)
    Kxx = rbf_kernel(x, x, gamma, c)

    mu = Kxxt @ Kinv @ yt
    cov = Kxx - Kxxt @ Kinv @ Kxxt.T
    return mu, cov



for i, gamma in enumerate([0.1, 1.0, 10.0, 100.0]):
    for j, c in enumerate([0.1, 1.0, 10.0, 100.0]):
        K = rbf_kernel(x, x, gamma, c) + torch.eye(x.shape[0])*s2
        mu = torch.zeros_like(x)
        axs[i*2,j].plot(x, mu)
        axs[i*2,j].fill_between(x.flatten(), mu.flatten()-K.diag(), mu.flatten()+K.diag(), alpha=0.5)
        axs[i*2,j].set_xlim([-1, 1])
        axs[i*2,j].set_ylim([-3, 3])
        axs[i*2,j].set_title(f'GP prior (gamma={gamma}, c={c})')
        post_mu, post_cov = posterior_pred(x, xtrain, ytrain, gamma, c)
        axs[i*2+1,j].plot(x, post_mu)
        axs[i*2+1,j].fill_between(x.flatten(), post_mu.flatten()-post_cov.diag(), post_mu.flatten()+post_cov.diag(), alpha=0.5)
        axs[i*2+1,j].scatter(xtrain, ytrain, color='red', zorder=5)

        axs[i*2+1,j].set_xlim([-1, 1])
        axs[i*2+1,j].set_ylim([-3, 3])
        axs[i*2+1,j].set_title(f'GP posterior (gamma={gamma}, c={c})')

plt.tight_layout()
plt.show()

```

```python

# Define the kernel with parameters
class RBFKernel(torch.nn.Module):
    def __init__(self, gamma, c):
        super(RBFKernel, self).__init__()
        self.gamma = torch.nn.Parameter(gamma)
        self.c = torch.nn.Parameter(c)

    def forward(self, x1, x2):
        gamma = torch.exp(self.gamma)
        c = torch.exp(self.c)
        return (-gamma*(torch.cdist(x1, x2)**2)).exp()*c

# Define the negative log marginal likelihood
def neg_log_marginal_likelihood(model, x, y, sigma):
    K = model(x, x)
    L = torch.linalg.cholesky(K + sigma**2 * torch.eye(len(x)))
    alpha = torch.linalg.solve(L,y)
    det = torch.prod(torch.diag(L))**2
    nll = 0.5 * torch.log(det) + 0.5 * y.t() @ alpha
    return nll.squeeze()
```

```python
# Initialize the model and optimizer
kernel = RBFKernel(torch.tensor([1.0]), torch.tensor([1.0]))
sigma = torch.tensor([1.0])
optimizer = optim.SGD(kernel.parameters(), lr=0.001)

# Define the data
xtrain = torch.tensor(train_X, dtype=torch.float32)
ytrain = torch.tensor(train_y, dtype=torch.float32)
best_loss = np.inf
best_params = None

for _ in range(10):
    # Randomly initialize the hyperparameters
    for param in kernel.parameters():
        param.data = torch.rand_like(param)
    
    # Optimize the hyperparameters
    for _ in range(1_000):
        optimizer.zero_grad()
        loss = neg_log_marginal_likelihood(kernel, xtrain, ytrain, sigma)
        loss.backward()
        optimizer.step()
    # Save the best kernel
    if loss < best_loss:
        best_loss = loss
        best_params = [param.clone() for param in kernel.parameters()]
```

```python
def post_pred(x, xt, yt, gamma=10.0, c=1.0):
    kernel = RBFKernel(gamma, c)
    Kxxt = kernel(x, xt)
    Kxt = kernel(xt, xt) + torch.eye(xt.shape[0])*s2
    Kinv = torch.linalg.inv(Kxt)
    Kxx = kernel(x, x)

    mu = Kxxt @ Kinv @ yt
    cov = Kxx - Kxxt @ Kinv @ Kxxt.T
    return mu, cov
```

```python
x = torch.linspace(-1,1,100).reshape(-1,1)
xtest = torch.tensor(test_X, dtype=torch.float32)
ytest = torch.tensor(test_y, dtype=torch.float32)
post_mu, post_cov = post_pred(x, xtest, ytest, best_params[0], best_params[1])
plt.scatter(test_X, test_y, color='red')
plt.plot(x, post_mu.detach().numpy())
plt.fill_between(x.flatten(), (post_mu.flatten()-post_cov.diag()).detach().numpy(), (post_mu.flatten()+post_cov.diag()).detach().numpy(), alpha=0.5)

plt.plot()
```
